```sql
select *
 from `workspace`.`default`.`bright_coffee_shop_analysis_raw_file_csv` limit 100;


 select
     transaction_date as purchase_date,
       dayname(transaction_date) as day_name,
        monthname(transaction_date) as month_name,
       dayofmonth(transaction_date)as day_of_month,
       transaction_time as purchase_time,
       
 case
     when dayname(transaction_date) in ('Sun','Sat')then 'weekend'
     else 'weekday'
 end as day_classification,
case
     when date_format(transaction_time,'HH:mm:ss') between '05:00:00' and '08:59:59' then '01.rush hour'
     when date_format(transaction_time,'HH:mm:ss') between '09:00:00' and '11:59:59' then '02.midmorning'
     when date_format(transaction_time,'HH:mm:ss') between '12:00:00' and '15:59:59' then '03.afternoon'
     when date_format(transaction_time,'HH:mm:ss') between '16:00:00' and '18:00:00' then '04.rush hour'
     else '05.evening'
    
     end as time_buckets,
     ------count of ids
     count(distinct transaction_id) as number_of_sales,
     count(distinct product_id) as number_of_products,
     count(distinct store_id)as number_of_stores,
     

---Revenue----
sum(unit_price*transaction_qty) as total_revenue_per_day, 

case
    when sum(unit_price*transaction_qty) <=50 then '01.low_spend'
    when sum(unit_price*transaction_qty) between 51 and 100 then '02.medium_spend'
    else '03.high_spend'
end as spend_buckets,

----Categorical columns--
store_location,
product_category,
product_detail,
product_type

from workspace.default.bright_coffee_shop_analysis_raw_file_csv 

group by
       transaction_date,
       dayname(transaction_date),
       monthname(transaction_date),
       dayofmonth(transaction_date),
       transaction_time,
       case
     when dayname(transaction_date) in ('Sun','Sat')then 'weekend'
     else 'weekday'
 end,
case
     when date_format(transaction_time,'HH:mm:ss') between '05:00:00' and '08:59:59' then '01.rush hour'
     when date_format(transaction_time,'HH:mm:ss') between '09:00:00' and '11:59:59' then '02.midmorning'
     when date_format(transaction_time,'HH:mm:ss') between '12:00:00' and '15:59:59' then '03.afternoon'
     when date_format(transaction_time,'HH:mm:ss') between '16:00:00' and '18:00:00' then '04.rush hour'
     else '05.evening'
     end,
     
     store_location,
      product_category,
      product_detail,
      product_type
order by transaction_date
```
